```python
import pandas as pd
import plotly.graph_objs as go
```

```python
workingset_sizes = [1, 2, 3, 4, 5, 6, 7, 8]
mig_methods = ['sha1/HBFDP', 'sha1/XBZRLE_HBFDP','md5/HBFDP', 'md5/XBZRLE_HBFDP']
# mig_methods = ['Vanilla', 'XBZRLE', 'HBFDP', 'XBZRLE_HBFDP']
```

```python
def mig_time_plot(plot_column="Total-Time", y_title="Total Migration Time (s)"):
    mig_data = {}
    for mig_method in mig_methods:
        mig_data[mig_method] = {"data": [], "deviation": []}
        for workingset in workingset_sizes:
            workingset_mb = str(workingset*1024)
            # data = pd.read_csv(f'./data/{mig_method}/{workingset_mb}.csv')
            data = pd.read_csv(f'./sha1 vs md5/{mig_method}/{workingset_mb}.csv')
            if plot_column == "Total-Time" or plot_column == "Downtime":
                mig_data[mig_method]["data"].append(data[plot_column].mean()/1000)
                mig_data[mig_method]["deviation"].append(data[plot_column].std()/1000)
            else:
                mig_data[mig_method]["data"].append(data[plot_column].mean())
                mig_data[mig_method]["deviation"].append(data[plot_column].std())
                
    print(mig_data)

    data = pd.DataFrame({'Workingset Size (GB)': [ str(workingset) for workingset in workingset_sizes],'SHA1_XBZRLE_HBFDP': mig_data['sha1/XBZRLE_HBFDP']["data"],'SHA1_HBFDP': mig_data['sha1/HBFDP']["data"],'MD5_XBZRLE_HBFDP': mig_data['md5/XBZRLE_HBFDP']["data"],'MD5_HBFDP': mig_data['md5/HBFDP']["data"]})
    # data = pd.DataFrame({'Workingset Size (GB)': [ str(workingset) for workingset in workingset_sizes],'Vanilla': mig_data['Vanilla']["data"],'XBZRLE': mig_data['XBZRLE']["data"],'XBZRLE_HBFDP': mig_data['XBZRLE_HBFDP']["data"],'HBFDP': mig_data['HBFDP']["data"]})

    print(data)

    # trace1 = go.Bar(x=data['Workingset Size (GB)'], y=data['Vanilla'], name='Vanilla Pre-Copy', error_y=dict(type='data', array=mig_data['Vanilla']["deviation"]))
    # trace2 = go.Bar(x=data['Workingset Size (GB)'], y=data['XBZRLE'], name='XBZRLE Pre-Copy', error_y=dict(type='data', array=mig_data['XBZRLE']["deviation"]))
    trace3 = go.Bar(x=data['Workingset Size (GB)'], y=data['SHA1_XBZRLE_HBFDP'], name='SHA1_XBZRLE_HBFDP Pre-Copy', error_y=dict(type='data', array=mig_data['sha1/XBZRLE_HBFDP']["deviation"]))
    trace4 = go.Bar(x=data['Workingset Size (GB)'], y=data['SHA1_HBFDP'], name='SHA1_HBFDP Pre-Copy', error_y=dict(type='data', array=mig_data['sha1/HBFDP']["deviation"]))
    trace5 = go.Bar(x=data['Workingset Size (GB)'], y=data['MD5_XBZRLE_HBFDP'], name='MD5_XBZRLE_HBFDP Pre-Copy', error_y=dict(type='data', array=mig_data['md5/XBZRLE_HBFDP']["deviation"]))
    trace6 = go.Bar(x=data['Workingset Size (GB)'], y=data['MD5_HBFDP'], name='MD5_HBFDP Pre-Copy', error_y=dict(type='data', array=mig_data['md5/HBFDP']["deviation"]))

    layout = go.Layout(
                    xaxis=dict(title='Workingset Size (GB)'),
                    yaxis=dict(title=y_title),
                    barmode='group',
                    template='simple_white'
                    )

    # fig = go.Figure(data=[trace1, trace2, trace3, trace4], layout=layout)
    fig = go.Figure(data=[trace3, trace4, trace5, trace6], layout=layout)

    fig.update_layout(
        width=677,  # 0.45 text width
        height=500,
        font=dict(family="Arial", size=17.2, color="black"),
        margin=dict(t=15, r=5, l=70, b=45, autoexpand=True),
        yaxis_title_standoff=10,
        xaxis_title_standoff=8,
        legend=dict(
            orientation="h",
            yanchor="bottom",
            # Align center
            y=-0.5,
        ),
    )
    fig.write_image(f"./Workingset-Size-vs-{plot_column}.png",scale=4, width=677, height=500)

    fig.show()
```

```python
mig_time_plot()
```

```python
mig_time_plot("Downtime", "Downtime (s)")
```

```python
mig_time_plot("Total-Fake-Dirty-Pages", "Total Fake Dirty Pages")
```
