```python
import pandas as pd
import os
import plotly.graph_objects as go
```

```python
def plot_graph(data, ram_size):
    layout = go.Layout( xaxis=dict(title='Time (S)'), yaxis=dict(title='Bitrate (Mbits/s)'), template='simple_white' )
    
    trace1 = go.Scatter(x=data['Time'], y=data['Bitrate'], name='Bitrate', mode='lines')
    fig = go.Figure(data=[trace1], layout=layout)
    
    fig.write_image(f"./iperf_{ram_size}.png",scale=1, width=1920)
    
    fig.show()

for file in os.listdir(".\iperf"):
    if file.endswith(".csv"):
        print(f'Generated Graph For RAM {file.split("_")[1].strip("ram")} ')
        csv_data = pd.read_csv(f'S:\Python Virtual ENV\Chart Generation\iperf\\{file}')
        plot_graph(csv_data, file.split("_")[1].strip("ram"))
        print("\n")
```
